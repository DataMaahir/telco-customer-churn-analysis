# Identifying Drivers of Customer Churn and Revenue Impact in a Telecom Company

## 1. Executive Summary

Customer churn represents a major challenge for subscription-based businesses because losing customers directly impacts recurring revenue. This analysis investigates customer churn patterns within a telecommunications dataset to identify factors associated with customer attrition and highlight potential strategies for improving customer retention.

The dataset contains information on customer demographics, service subscriptions, billing behavior, and churn status. After cleaning and preparing the data, key performance indicators were calculated to assess the scale of churn within the customer base. Approximately 26–27% of customers in the dataset have churned, representing a significant proportion of the total customer base.

Exploratory data analysis revealed several factors strongly associated with churn. Customers on month-to-month contracts exhibit substantially higher churn compared to customers on longer-term contracts. Additionally, churn is concentrated among customers with shorter tenure, indicating that many customers leave during the early stages of their relationship with the company.

The analysis also suggests that customers with higher monthly charges tend to churn slightly more often, while fiber optic internet users exhibit higher churn levels compared to DSL users and customers without internet service. Furthermore, customers paying through electronic checks show noticeably higher churn compared to customers using automated payment methods such as credit cards or bank transfers.

These findings suggest several opportunities for reducing churn, including encouraging customers to adopt longer-term contracts, improving onboarding and early customer engagement, promoting automated billing methods, and evaluating the customer experience associated with premium service plans.

Overall, this analysis provides actionable insights that can support data-driven strategies for improving customer retention and reducing customer attrition.

## 4. KPI Definitions

To evaluate customer churn and its business impact, the following Key Performance Indicators (KPIs) are defined for this analysis.

These KPIs provide a structured way to measure customer attrition, identify high-risk segments, and estimate the financial impact of churn.

### 4.1 Churn Rate (Primary KPI)

**Definition:**
Churn rate represents the percentage of customers who have discontinued their service during the observed period.

**Calculation:**

$$
\text{Churn Rate} =
\frac{\text{Number of Churned Customers}}
{\text{Total Customers}} \times 100
$$

**Business Importance:**
Churn rate is the primary indicator of customer retention. A high churn rate suggests customer dissatisfaction and directly impacts recurring revenue.

### 4.2 Total Customers

**Definition:**
Total number of customers included in the dataset.

**Calculation:**

$$
\text{Total Customers} = \text{Total Number of Records}
$$

**Business Importance:**
This KPI represents the size of the customer base and provides context for interpreting churn levels.

### 4.3 Total Churned Customers

**Definition:**
Total number of customers who have discontinued service.

**Calculation:**

$$
\text{Total Churned Customers} =
\text{Count of Customers Where Churn = Yes}
$$

**Business Importance:**
This KPI measures the absolute scale of customer attrition and complements churn rate analysis.

### 4.4 Revenue Lost Due to Churn

**Definition:**
Estimated monthly revenue associated with customers who have churned.

**Calculation:**

$$
\text{Revenue Lost} =
\sum(\text{Monthly Charges of Churned Customers})
$$

**Business Importance:**
This KPI estimates the financial impact of customer churn and helps quantify the potential benefits of retention strategies.

### 4.5 Average Revenue Per Customer (ARPU)

**Definition:**
Average monthly revenue generated per customer.

**Calculation:**

$$
\text{ARPU} =
\frac{\text{Total Monthly Charges}}
{\text{Total Customers}}
$$

**Business Importance:**
ARPU provides insight into customer value and helps identify whether high-value customers are more likely to churn.

### 4.6 Average Tenure of Churned Customers

**Definition:**
Average duration of service for customers who have churned.

**Calculation:**

$$
\text{Average Tenure (Churned)} =
\text{Mean Tenure of Churned Customers}
$$

**Business Importance:**
This KPI helps identify whether customers tend to leave early or after long-term usage, which is important for designing retention strategies.

### 4.7 Segment-Level Churn Rate

**Definition:**
Churn rate calculated separately for different customer segments such as contract type, payment method, or tenure group.

**Calculation Example:**

$$
\text{Segment Churn Rate} =
\frac{\text{Churned Customers in Segment}}
{\text{Total Customers in Segment}} \times 100
$$

**Business Importance:**
Segment-level churn rates help identify high-risk customer groups and enable targeted retention strategies.

## 6. Data Cleaning

### 6.1 Missing Values

The dataset was examined for missing values to ensure data completeness before analysis.
The first step is to check for explicit missing values as done below.

```python
df.isnull().sum() # Check for explicit missing values
```

No explicit missing value have been detected.

However, missing values can hide in the form of blank strings, or strings like "NaN", "Null", etc.

The **TotalCharges** column which is tenure(numeric) multiplied by MonthlyCharges(numeric) is of the string data type, when it should be of the type float(numeric).

The column was therefore tested for hidden missing values by attempting conversion to numeric format.

```python
pd.to_numeric(df['TotalCharges'], errors='coerce').isnull().sum()
```

The conversion revealed 11 invalid entries in the **TotalCharges** column. These values could not be converted to numbers and were interpreted as missing values.

These entries were originally stored as blank strings, which prevented them from being detected during the initial missing value check.

Since the missing values represent less than 0.2% (11/7043) of the dataset, removing these records is unlikely to affect the overall analysis.

Therefore, rows containing missing values in the TotalCharges column were removed.

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce') # Converting from string to numeric
```

```python
df = df.dropna(subset=['TotalCharges']) # Dropping missing values
```

```python
df.isnull().sum() # Rechecking for missing values after alteration
```

## 7. KPI Calculation

Key Performance Indicators (KPIs) were calculated to quantify the scale of customer churn and evaluate its financial impact.

These metrics provide a high-level overview of customer retention, revenue exposure, and customer behavior.

```python
total_customers = len(df)

churned_customers = (df['Churn'] == 'Yes').sum()

churn_rate = churned_customers / total_customers * 100

revenue_lost = df[df['Churn'] == 'Yes']['MonthlyCharges'].sum()

arpu = df['MonthlyCharges'].mean()

avg_churn_tenure = df[df['Churn'] == 'Yes']['tenure'].mean()
```

```python
churn_rate = round(churn_rate, 2)
revenue_lost = round(revenue_lost, 2)
arpu = round(arpu, 2)
avg_churn_tenure = round(avg_churn_tenure, 2)
```

```python
kpi_table = pd.DataFrame({
    "KPI": [
        "Total Customers",
        "Total Churned Customers",
        "Churn Rate (%)",
        "Revenue Lost (Monthly)",
        "Average Revenue Per Customer (ARPU)",
        "Average Tenure of Churned Customers"
    ],
    "Value": [
        total_customers,
        churned_customers,
        churn_rate,
        revenue_lost,
        arpu,
        avg_churn_tenure
    ]
})
kpi_table
```
